## Combine Data

```python
import pandas as pd

def load_and_combine_data(artist_path, artist_tracks_path, tracks_path, youtube_path):
    # Load each CSV into a DataFrame
    artist_df = pd.read_csv(artist_path)
    artist_tracks_df = pd.read_csv(artist_tracks_path)
    tracks_df = pd.read_csv(tracks_path)
    youtube_df = pd.read_csv(youtube_path)

    # Print data types for verification
    # print("Tracks DataFrame data types:")
    # print(tracks_df.dtypes)
    # print("YouTube DataFrame data types:")
    # print(youtube_df.dtypes)
 
    # Convert ID columns to string to ensure consistency
    artist_df['id'] = artist_df['id'].astype(str)
    artist_tracks_df['cm_artist'] = artist_tracks_df['cm_artist'].astype(str)
    tracks_df['cm_track'] = tracks_df['cm_track'].astype(str)
    artist_tracks_df['cm_track'] = artist_tracks_df['cm_track'].astype(str)
    youtube_df['cm_track'] = youtube_df['cm_track'].astype(str)

    # Merge data with unique suffixes for overlapping columns
    combined_df_1 = pd.merge(artist_df, artist_tracks_df, left_on='id', right_on='cm_artist', how='left', suffixes=('', '_src_artist_track'))
    combined_df_2 = pd.merge(combined_df_1, tracks_df, on='cm_track', how='left', suffixes=('', '_src_track'))
    final_combined_df = pd.merge(combined_df_2, youtube_df, left_on='cm_track', right_on='cm_track', how='left', suffixes=('', '_src_youtube'))

    return final_combined_df

def main():
    try:
        # Define paths to the CSV files
        artist_path = 'artist_df.csv'  # Adjust these paths if needed
        artist_tracks_path = 'artist_tracks_df.csv'
        tracks_path = 'tracks_df.csv'
        youtube_path = 'youtube_df.csv'
        
        # Print to verify paths are correct
        print(f"Artist path: {artist_path}")
        print(f"Artist tracks path: {artist_tracks_path}")
        print(f"Tracks path: {tracks_path}")
        print(f"YouTube path: {youtube_path}")

        # Load and combine data from all CSV files
        print("Loading data...")
        combined_data = load_and_combine_data(artist_path, artist_tracks_path, tracks_path, youtube_path)
        
        print("Data loaded successfully. Here's a preview:")
        print(combined_data.head())  # Display the first few rows to check
        
        # Check if DataFrame is empty
        if combined_data.empty:
            print("No data available in DataFrame.")
        else:
            # Save the combined DataFrame to a new CSV file
            output_path = 'combined_data.csv'
            combined_data.to_csv(output_path, index=False)
            print(f"Data combined and saved to {output_path}.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Call the main function
main()
```
